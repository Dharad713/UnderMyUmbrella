```python
import os
from dotenv import load_dotenv

load_dotenv()

latitude = float(os.getenv("HOME_LAT"))
longitude = float(os.getenv("HOME_LONG"))
```

```python
import requests
url = "https://api.open-meteo.com/v1/forecast"

params = {
    "latitude": latitude,
    "longitude": longitude,
    "daily": "weather_code,temperature_2m_max,temperature_2m_min,precipitation_probability_max",
    "temperature_unit": "fahrenheit",
    "timezone": "America/New_York",
    "forecast_days": 3,
    "past_days": 3
}

headers = {
    "Accept": "application/json",
    "User-Agent": "UnderMyUmbrella/1.0",
    "Accept-Encoding": "gzip, deflate",
}

response = requests.get(url, params=params, headers=headers, timeout=10)
response.raise_for_status()

print(response.json())
forecast = response.json()["daily"]

print("Date:", forecast["time"][0])
print("High:", forecast["temperature_2m_max"][0], "°F")
print("Low:", forecast["temperature_2m_min"][0], "°F")
print(
    "Rain chance:",
    forecast["precipitation_probability_max"][0],
    "%",
)
```
